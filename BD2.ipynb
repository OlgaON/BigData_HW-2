```
from pyspark.sql.functions import mean, stddev, variance, col
```

```
df_state=df.select(
mean(col("SalePrice")).alias("mean"),
stddev(col("SalePrice")).alias("std"),
variance(col("SalePrice")).alias("std")
).collect()
```
